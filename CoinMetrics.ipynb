```
coinmetrics_start = 'https://community-api.coinmetrics.io/v4/'
coinmetrics_func = 'catalog/metrics?pretty=true'
# coinmetrics_func = 'timeseries/asset-metrics?assets=btc&metrics=PriceUSD,TxCnt&frequency=1d'
# coinmetrics_func = 'timeseries/asset-metrics?assets=btc&metrics=TxCnt&frequency=1d'

# queryStringParams = {
#   'vs_currency':'usd',
#   'days':'.9'
# }
# encodedParams = urllib.parse.urlencode(queryStringParams)

# https://stackoverflow.com/questions/31778800/how-can-i-make-a-post-request-on-python-with-urllib3
# https://stackoverflow.com/questions/7933417/how-do-i-set-headers-using-pythons-urllib
# https://www.pluralsight.com/guides/importing-data-from-json-resource-with-python

with request.urlopen(coinmetrics_start + coinmetrics_func) as response:
        if response.getcode() == 200:
            source = response.read()
            data = json.loads(source)
        else:
            print('An error occurred while attempting to retrieve data from the API.')
```

```
# https://coinmetrics.readthedocs.io/en/latest/usage.html

# Initialize a reference object, in this case `cm` for the Community API
cm = coinmetrics.Community()

# Usage Examples ############################################################

# List the assets Coin Metrics has data for.
supported_assets = cm.get_supported_assets()
print("supported assets:\n", supported_assets)

len(supported_assets)

# List all available metrics for BTC.
#asset = "btc"
asset = "eth"
available_data_types = cm.get_available_data_types_for_asset(asset)
print("available data types:\n", available_data_types)

# Fetch the `PriceUSD` and `ROI30d` data for BTC from 2019-01-01 to 2019-01-08.
#asset = "btc"
#asset = "ltc"
#metric = "PriceUSD,ROI30d,TxCnt,FeeMedUSD,FeeTotUSD,CapMrktFFUSD,DiffLast,HashRate,TxTfrValAdjNtv"
#metric = "PriceUSD,ROI30d,TxCnt,FeeMedUSD,FeeTotUSD,CapMrktFFUSD"

metric = ','.join(available_data_types) # all available data for coin

begin_timestamp = "2010-01-01"  # The `datetime` type is also accepted
end_timestamp = "2021-09-22"  # The `datetime` type is also accepted
asset_data = cm.get_asset_data_for_time_range(asset, metric, begin_timestamp, end_timestamp)
#print("data given timerange:\n", asset_data)
print('done')
```

```
# Convert the data object we recieved to a Pandas DataFrame for further processing.
# We are reusing the `asset_data` from the previous step.
df = coinmetrics.cm_to_pandas(asset_data)

df.index = pd.to_datetime(df.index)

df
```

```
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

print(df.shape)

plt.figure(figsize=(12,8))
plt.title('BTC Historical Price USD ($)')
plt.xlabel('Date')
plt.ylabel('Close Price USD ($')
plt.plot(df.PriceUSD)
plt.show()
```

```
price_df = df.PriceUSD.dropna().reset_index()
#print(price_df.shape)
# hello

future_days = 200
price_df['Prediction'] = price_df[['PriceUSD']].shift(-200)
#print(price_df.shape)
#price_df

X = np.array(price_df[['PriceUSD']])[:-future_days]
y = np.array(price_df.Prediction)[:-future_days]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8)
print(X_train.shape)
#print(y_train.shape)
print(X_test.shape)
#print(y_test.shape)
```

```
linreg = LinearRegression().fit(X_train, y_train)
linreg_prediction = linreg.predict(X_test)

#print(linreg_prediction)

valid = pd.DataFrame(y_test)
valid['pred'] = linreg_prediction
valid 
```
